```python
Q1= np.percentile(df['feature_1'],.25, method='midpoint')
Q2= np.percentile(df['feature_1'],.50, method='midpoint')
Q3= np.percentile(df['feature_1'],.75, method='midpoint')

print(Q1,Q2,Q3)
```

```python
IQR=Q3-Q1
Q3
```

```python
upper_limit=Q3+1.5*IQR
upper_limit
```

```python
lower_limit=Q1-1.5*IQR
lower_limit
```

```python
outliers=[]
for x in df['feature_1']:
    if((x > upper_limit) or (x < lower_limit)):
        outliers.append(x)
outliers
```
