```python
for i in ['Engine HP','Engine Cylinders','Number of Doors']:
    df[i]=df[i].fillna(df[i].median())
df[i]
```

```python
Q1=np.percentile(df['Engine HP'], 25, method='midpoint')
Q2=np.percentile(df['Engine HP'], 50, method='midpoint')
Q3=np.percentile(df['Engine HP'], 75, method='midpoint')

print(Q1,Q2,Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
ulimit=Q3+1.5*IQR
llimit=Q1-1.5*IQR
print(ulimit,llimit)
```

```python
outliers=[]
for x in df['Engine HP']:
    if((x > ulimit) or (x < llimit)):
        outliers.append(x)
outliers
```

```python
df2=df['Engine HP'] > ulimit
df2
```

```python
df.loc[df2]
```

```python
a=df.index[df2]
```

```python
df.drop(a, inplace=True)
```
